```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airbnb.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
without_name = df.drop(['name','host_name','host_id','neighbourhood'], axis=1)
```

```python
without_name.head()
```

```python
without_name.isnull().sum()
```

```python
mean = without_name["reviews_per_month"].mean()
filling1 = without_name["reviews_per_month"].fillna(mean)
filling2 = without_name["last_review"].fillna(df["last_review"].mode()[0])
```

```python
without_name["reviews_per_month"] = filling1
without_name["last_review"] = filling2
```

```python
without_name.isnull().sum()
```

```python
data_cleaned = without_name
data_cleaned.head()
```

```python
data_cleaned.describe(include='all')
```

```python
condition = [
            (data_cleaned["price"] <= 30),
            (data_cleaned["price"] > 30) & (df["price"] <= 70),
            (data_cleaned["price"] > 70) 
            ]
decision = ['Cheap', 'Medium', 'Expensive']
data_cleaned["category"] = np.select(condition, decision)
```

```python
data_cleaned.head()
```

```python
plt.hist(data_cleaned['price'], 30, range=[0, 400], facecolor='gray', align='mid')
plt.title("Harga dari Kamar yang Tersedia")
plt.xlabel("Harga")
plt.show()
```
